```python
import nltk
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
text = """
Artificial Intelligence is changing the world.
Machines can now understand language, recognize images,
and even chat with humans intelligently.
"""
```

```python
import nltk

nltk.download('punkt')
nltk.download('stopwords')
```

```python
nltk.download('punkt_tab')
```

```python
tokens = word_tokenize(text)

print("Original Tokens:")
print(tokens)
```

```python
['Artificial', 'Intelligence', 'is', 'changing', 'the',
'world', '.', 'Machines', 'can', 'now', 'understand',
'language', ',', 'recognize', 'images', ',', 'and',
'even', 'chat', 'with', 'humans', 'intelligently', '.']
```

```python
stop_words = set(stopwords.words('english'))

filtered_tokens = []

for word in tokens:
    word_lower = word.lower()

    if word_lower not in stop_words and word not in string.punctuation:
        filtered_tokens.append(word_lower)

print("After Removing Stopwords & Punctuation:")
print(filtered_tokens)
```

```python
['artificial', 'intelligence', 'changing', 'world',
'machines', 'understand', 'language',
'recognize', 'images', 'even',
'chat', 'humans', 'intelligently']
```

```python
bow = Counter(filtered_tokens)

print("Bag of Words Representation:")
print(bow)
```
