```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sentences = [
    "I love playing football",
    "Soccer is my favorite sport",
    "The weather is very hot today",
    "Artificial Intelligence is changing the world",
    "Machine learning powers modern AI systems"
]

print(sentences)
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')

embeddings = model.encode(sentences)

print("Embedding generated successfully!")
```

```python
similarity_1 = cosine_similarity(
    [embeddings[0]],
    [embeddings[1]]
)

similarity_2 = cosine_similarity(
    [embeddings[0]],
    [embeddings[2]]
)

similarity_3 = cosine_similarity(
    [embeddings[3]],
    [embeddings[4]]
)

print("Football vs Soccer:", similarity_1[0][0])
print("Football vs Weather:", similarity_2[0][0])
print("AI vs Machine Learning:", similarity_3[0][0])
```
