```python
import time

# Test Queries
test_queries = [
    "What is AI?",
    "What is Machine Learning?",
    "What is Deep Learning?",
    "What is NLP?",
    "What is RAG?"
]

# Dummy RAG Function
# Isko baad mein apne actual RAG code se replace karna
def timed_rag(query):

    timings = {}

    start = time.perf_counter()

    # Embedding
    t1 = time.perf_counter()
    time.sleep(0.1)  # simulate embedding
    timings["embedding"] = time.perf_counter() - t1

    # FAISS Search
    t2 = time.perf_counter()
    time.sleep(0.05)
    timings["faiss_search"] = time.perf_counter() - t2

    # Prompt Construction
    t3 = time.perf_counter()
    time.sleep(0.02)
    timings["prompt_construction"] = time.perf_counter() - t3

    # LLM Generation
    t4 = time.perf_counter()
    time.sleep(0.3)
    timings["llm_generation"] = time.perf_counter() - t4

    # Formatting
    t5 = time.perf_counter()
    result = {
        "answer": f"Answer for {query}",
        "sources": ["doc1", "doc2"]
    }
    timings["formatting"] = time.perf_counter() - t5

    timings["total"] = time.perf_counter() - start

    return result, timings


# Run Profiling
all_timings = []

for query in test_queries:

    result, timing = timed_rag(query)

    all_timings.append(timing)

    print(f"\nQuery: {query}")
    print("Answer:", result["answer"])
    print("Timing:", timing)


# Calculate Averages
steps = [
    "embedding",
    "faiss_search",
    "prompt_construction",
    "llm_generation",
    "formatting"
]

print("\n===== AVERAGE TIMINGS =====")

avg = {}

for step in steps:

    avg[step] = sum(t[step] for t in all_timings) / len(all_timings)

    print(f"{step}: {avg[step]:.4f} sec")


# Percentage Breakdown
total_avg = sum(avg.values())

print("\n===== PERCENTAGE BREAKDOWN =====")

for step in steps:

    percentage = (avg[step] / total_avg) * 100

    print(f"{step}: {percentage:.2f}%")
```
