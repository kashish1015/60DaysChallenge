```python
import os
import re
import faiss
import numpy as np
import pandas as pd
from tabulate import tabulate
from openai import OpenAI
```

```python
def get_embedding(text):

    response = client.embeddings.create(
        model="text-embedding-3-small",
        input=text
    )

    return response.data[0].embedding
```

```python
print("Generating embeddings...")

embeddings = []

for doc in corpus:
    vector = get_embedding(doc)
    embeddings.append(vector)

embeddings = np.array(embeddings).astype("float32")

print("Embeddings shape:", embeddings.shape)
```

```python
dimension = embeddings.shape[1]

index = faiss.IndexFlatL2(dimension)

index.add(embeddings)

print("FAISS index size:", index.ntotal)
```

```python
def semantic_search(query, top_k=3):

    query_vector = np.array(
        [get_embedding(query)],
        dtype="float32"
    )

    distances, indices = index.search(query_vector, top_k)

    results = []

    for rank, idx in enumerate(indices[0]):

        results.append({
            "document": corpus[idx],
            "distance": float(distances[0][rank])
        })

    return results
```

```python
def tokenize(text):
    return re.findall(r"\w+", text.lower())


def keyword_search(query, corpus, top_k=3):

    query_words = set(tokenize(query))

    scores = []

    for doc in corpus:

        doc_words = set(tokenize(doc))

        overlap = len(query_words.intersection(doc_words))

        scores.append((doc, overlap))

    scores.sort(key=lambda x: x[1], reverse=True)

    return scores[:top_k]
```

```python
queries = [
    "medicine for sick people",
    "space travel company",
    "ways to reduce stress",
    "computer hacking protection",
    "healthy exercise activities",
    "financial investment startup",
    "movies about future worlds",
    "fixing software problems",
    "cars without gasoline",
    "books and reading places"
]
```

```python
comparison_rows = []

for query in queries:

    semantic_results = semantic_search(query, top_k=1)
    keyword_results = keyword_search(query, corpus, top_k=1)

    comparison_rows.append([
        query,
        semantic_results[0]["document"],
        keyword_results[0][0]
    ])

df = pd.DataFrame(
    comparison_rows,
    columns=[
        "Query",
        "Semantic Search Top Result",
        "Keyword Search Top Result"
    ]
)

print(tabulate(df, headers="keys", tablefmt="grid"))
```

```python
query = "space exploration company"

print("QUERY:", query)

print("\nSemantic Search Results:\n")

for result in semantic_search(query):

    print(result)

print("\nKeyword Search Results:\n")

for result in keyword_search(query, corpus):

    print(result)
```

```python
# Trade-Off Analysis

## Semantic Search Advantages
- Understands meaning and intent
- Handles synonyms and paraphrases
- Better for natural language queries

## Semantic Search Disadvantages
- More expensive
- Higher latency
- Requires embeddings

## Keyword Search Advantages
- Fast and cheap
- Exact matching
- Easy to debug

## Keyword Search Disadvantages
- Fails on vocabulary mismatch
- Cannot understand intent

## Production Scale Considerations
- FAISS scales efficiently
- Flat indexes work for small datasets
- IVF/HNSW indexes are better for millions of vectors
```
